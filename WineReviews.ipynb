# 3. 数据缺失的处理

```python
data.isnull().sum()
```

存在数据缺失的属性有：country、designation、price、province、region_1、region_2，数据缺失的原因可能是由于忘记对其进行填写

### country和province

由于country和province缺失的数据极少，所以采用将缺失部分剔除的方法

```python
data_dropna = data.dropna(subset=['country', 'province'])
```

对比新旧数据集:

```python
plt.figure(figsize=(20,20))
plt.subplot(2,1,1) 
data['country'].value_counts(dropna = False).plot(kind="bar")
plt.subplot(2,1,2)
data_dropna['country'].value_counts(dropna = False).plot(kind="bar")
```

```python
plt.figure(figsize=(20,20))
plt.subplot(2,1,1) 
data['province'].value_counts(dropna = False)[:100].plot(kind="bar")
plt.subplot(2,1,2)
data_dropna['province'].value_counts(dropna = False)[:100].plot(kind="bar")
```

### region_1和designation

region_1的数据比较分散，并不集中在某些值上。同时region_1表示的是地址，难以利用属性的相关关系进行推断。所以也采用将缺失部分剔除的方法对它进行处理。

```python
data_dropna = data_dropna.dropna(subset=['region_1', 'designation'])
```

对比新旧数据集:

```python
plt.figure(figsize=(20,20))
plt.subplot(2,1,1) 
data['region_1'].value_counts(dropna = False)[:100].plot(kind="bar")
plt.subplot(2,1,2)
data_dropna['region_1'].value_counts(dropna = False)[:100].plot(kind="bar")
```

```python
plt.figure(figsize=(20,20))
plt.subplot(2,1,1) 
data['designation'].value_counts(dropna = False)[:100].plot(kind="bar")
plt.subplot(2,1,2)
data_dropna['designation'].value_counts(dropna = False)[:100].plot(kind="bar")
```

### price

由于在price中，大量的数据都集中在最高频率值对应的数据，所以采用用最高频率值来填补缺失值的方法

```python
price_fillna = data_dropna['price'].fillna(data['price'].mode()[0])
```

对比新旧数据集:

```python
plt.figure(figsize=(20,20))
plt.subplot(1,2,1) 
data.boxplot(['price'])
plt.subplot(1,2,2) 
pd.DataFrame(price_fillna).boxplot()
```
